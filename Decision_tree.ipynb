# Дерево решений

Вместо тысячи слов один пример:

В деревьях решений есть несколько ключевых параметров:  
1. Количество признаков, по которым строится дерево.  
2. Ветвистость дерева (количество условий, по которым идет классификация)  

Основная проблема  - это переобучить деревья на обучающей выборке или недообучить.   
Для оптимальной глубины дерева используется кросс-валидация. (Кросс-валидацией называется метод обучения и оценки модели, при котором данные разбиваются на несколько частей)

```python
from sklearn import tree


X = [[1,2], [2,3], [1,3], [10,12], [11,13],[13,10]]
y = [1,1,1,0,0,0]
clr = tree.DecisionTreeClassifier()
```

```python
clr.fit(X, y) # X - это входные данные (предикторы), y - это выходные значения (целевые значения)
```

```python
print(clr.predict([[10,10]]))
print(clr.predict([[3,1]]))
```

Следующая функция показывает, как строится данное дерево решений

```python
tree.plot_tree(clr) 
```

Рассмотрим более сложный пример построения дерева решений на наборе ирисов

```python
from sklearn.datasets import load_iris
from sklearn import tree

iris = load_iris()
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,y)
```

```python
tree.plot_tree(clf)
```
