# **Линейная регрессия**

## **Теоритическая часть**

Линейная регрессия  - это модель зависимости переменной y от переменных (x1,x2,...,xn), где n - любое натуральное число.

Уравнение выглядит в виде: yi = a0 + ai*xi, где yi,xi - известные значения. Необходимо найти a и b.

Другими словами, задача провести прямую между точками так, чтобы точки лежали максимально плотно к этой прямой. Пример ниже:


изображение.png

Почему же кратчайшее расстояние определяется как перпендикуляр опущенный на прямую относительно оси Ox? Почему бы перпендикуляр относительно Oy или же просто к прямой не опустить?
Можно и так, но это будет другое уравнение и другой раздел (факторный анализ)

Определения:

*   Y - отклик / зависимая переменная
*   X - предиктор / независимая переменная
*   epsilon_i = yi - (a+bxi) - невязка / ошибка







Как понять что модель работает хорошо / плохо? Для этого используются критерии качества. Самые популярные из них:
1.   MSE = 1/n*SSE - среднее суммы квадратов ошибок
2.   SSE - сумма квадратов ошибок

Задача линейной регрессии в ML - минимизация критерия качества. Т.е. MSE -> min


Почему мы уверенны, что при минимизации MSE получим прямую, вокруг которой точки лежат наиболее плотно? Данное утверждение несложно проверить через функцию правдоподобия и нормальное Гауссовское распределение. Если взять тот факт, что матожидание М =0, а дисперсия (sigma)^2, т.е. N(0,(sigma^2)) и подставить в функцию правдоподобия, то в результате минимизации получим среднее суммы квадратов ошибок :)

## **Практическая часть**

```python
from sklearn.datasets import load_iris
import pandas as pd

def sklearn_to_df(sklearn_dataset):
    df = pd.DataFrame(sklearn_dataset.data, columns=sklearn_dataset.feature_names)
    df['target'] = pd.Series(sklearn_dataset.target)
    return df

df = load_iris()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X = df.data
Y = df.target
X_train,X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("MSE:", mse)
print("R-squared", r2)
```
